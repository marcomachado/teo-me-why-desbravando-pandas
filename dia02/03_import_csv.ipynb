```python
# quem está entre 1000 e 2000 pontos
condicao_pontos = (df_customer['Points'] >= 1000) & (df_customer['Points'] <= 2000)
df_1000_2000 = df_customer[condicao_pontos]
df_1000_2000.head()
```

```python
df_1000_2000.describe()
```

```python
# Python só trabalha com referências
a = [1,2,3,4]
b = a # lista B aponta para referência da lista A
print(a)
print(b)
c = a.copy() # copia (cria nova lista) valores e novas referências

b.append(5) #add valor 5 na lista B (que possui mesmas referencias da lista A)
print(a)
print(b)
print(c)
```

```python
## cada filtro no DF, NÃO cria um novo DF
## os "novos DFs" contém referências para o original
## se tentar modificar um DF fatiado (filtrado) -> alerta de que pode estar modificando dados originais
## está tentando modificar a partir de uma referência

df_1000_2000['Points'] = df_1000_2000['Points'] + 1000
```

```python
# se precisa manipular os dados -> fazer cópia
df_1000_2000 = df_customer[condicao_pontos].copy()
df_1000_2000['Points'] = df_1000_2000['Points'] + 1000
df_1000_2000.head()
```
