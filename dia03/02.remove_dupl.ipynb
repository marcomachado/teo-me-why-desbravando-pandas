```python
import pandas as pd
```

```python
data = {
    "Nome": ["Téo", "Nah", "Maria", "Nah", "Lara", "Téo"],
    "Idade": [32,33,2,33,31,32]
}

df = pd.DataFrame(data)
df
```

```python
df.drop_duplicates()
```

```python
data2 = {
    "Nome": ["Téo", "Nah", "Maria", "Nah", "Lara", "Téo"],
    "Idade": [32,33,2,33,31,32],
    "updated_at":[1,2,3,1,2,3]
}
df2 = pd.DataFrame(data2)
df2
```

```python
# todas as linhas são diferentes -> definir critério para considerar duplicado
# mantem primeiro registro; apaga os outros
df2.drop_duplicates(subset=['Nome', 'Idade'])
```

```python
# keep = 'first' (default) mantem na lista o primeiro que encontrar
# problema: manter o último registro (mais atualizado > updated_at)

# 1ª solucao: manter ultimos
# ha chance de erros se o DF nao tiver criterio de ordenacao nas linhas, como abaixo
# manteve Nah com updated_at 1 e apagou updated_at = 2

df2.drop_duplicates(subset=['Nome', 'Idade'], keep='last')
```

```python
#2ª solucao: ordenar e depois usa drop default
df2 = df2.sort_values(by='updated_at', ascending=False).drop_duplicates(subset=['Nome', 'Idade'])
df2
```
