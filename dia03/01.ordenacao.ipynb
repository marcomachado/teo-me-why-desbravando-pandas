```python
import pandas as pd
```

```python
df = pd.read_csv('../data/customers.csv', sep=';')

df
```

```python
df.sort_values(by="Points", ascending=False)
```

```python
df.sort_values(by="Points", ascending=False, inplace=True)
df.head()
```

```python
# sem inplace -> retorna DF para ser renomeado
df = df.sort_values(by="Points", ascending=False).rename(columns={
    'Name':'Nome',
    'Points':'Pontos'
})
df.tail(10)
```

```python
# ordenação por Pontos -> DESC / Nome -> ASC
df.sort_values(by=['Pontos', 'Nome'], ascending=[False, True], inplace=True)
df.tail(10)
```
