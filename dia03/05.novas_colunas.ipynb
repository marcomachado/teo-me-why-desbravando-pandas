```python
# Como usar apply em um DataFrame (usar mais uma coluna para processamento)

# simulando CRM
# Teo comprou a 1 dia / valor total gasto=100 / comprou 2x
# Nah comprou a 30 dias / valor total = 2000 / comprou 5x

# combinar recencia, valor, frequencia -> gerar uma nota

data = {
    "nome": ["Teo", "Nah", "Maria", "Lara"],
    "recencia": [1,30,10,45],
    "valor":[100,2000, 15, 500],
    "frequencia":[2, 5, 1, 15]
}

df_crm = pd.DataFrame(data)
df_crm
```

REGRAS DO CRM PARA CÁLCULO DE PONTOS (Soma)

|                |            |                      |           |
|----------------|------------|----------------------|-----------|
| **recencia**   | **< 10**   | **entre 10 e 30**    | **> 30**  |
|                | 10         | 5                    | 0         |
| **valor**      | **> 1000** | **entre 1000 e 100** | **< 100** |
|                | 10         | 5                    | 0         |
| **frequencia** | **> 10**   | **entre 10 e 5**     | **< 5**   |
|                | 10         | 5                    | 0         |

```python
def rfv(row):
    
    nota = 0
    
    if row['recencia'] <= 10:
        nota += 10
    elif 10 < row['recencia'] <= 30:
        nota += 5
    elif row['recencia'] > 30:
        nota += 0

    if row['valor'] > 1000:
        nota += 10
    elif 100 <= row['valor'] < 1000:
        nota += 5
    elif row['valor'] < 100:
        nota += 0

    if row['frequencia'] > 10:
        nota += 10
    elif 5 <= row['frequencia'] < 10:
        nota += 5
    elif row['frequencia'] < 5:
        nota += 0
    
    return nota
```

```python
# retorna uma Série com valor da primeira coluna
df_crm.iloc[0]
```

```python
df_crm['Pontos_RFV'] = df_crm.apply(rfv, axis=1)
df_crm
```
