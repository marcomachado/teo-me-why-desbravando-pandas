1. Converta a seguinte lista de dados para uma Series Pandas e obtenha:
 * Média
 * Desvio Padrão
 * Máximo Valor

dados = [10, 20, 42, 9, 12, 35, 24, 10, 8, 14, 21]


```python
import pandas as pd
```

```python
dados = [10, 20, 42, 9, 12, 35, 24, 10, 8, 14, 21]
```

```python
series_pd = pd.Series(dados)
```

```python
media = series_pd.mean()
media
```

```python
desvio = series_pd.std()
desvio
```

```python
maximo = series_pd.max()
maximo
```
