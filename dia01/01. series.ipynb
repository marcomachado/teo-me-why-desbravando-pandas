```python
import pandas as pd
```

```python
idades = [30, 42, 90, 34]
idades
```

```python
media = sum(idades) / len(idades)
media
```

```python
total = 0

for i in idades:
  total += (media-i)**2

variancia = total/(len(idades) - 1)
variancia
```
