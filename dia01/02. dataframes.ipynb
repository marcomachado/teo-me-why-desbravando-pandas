```python
import pandas as pd
```

```python
# dicionário
data = {
    "nome":["teo", 'nah', "lara", "maria"],
    "sobrenome":["calvo", "machado", "santos", "oliveira"],
    "idade": [30, 31, 32, 33]
}
```

```python
data
```

```python
# idade da primeira pessoa
data['idade'][0]
```

```python
# idade da última pessoa
data['idade'][-1]
```

```python
df = pd.DataFrame(data)
df
```

```python
# retorna uma Series
df['idade']
```

```python
# mostrar idade da primeira pessoa
print(df['idade'][0]) #se tiver certeza que o índice está correto
print(df['idade'].iloc[0]) # usar iloc garante que vem o primeiro elemento
```

```python
# DataFrame é um conjunto de Series

type(df['idade'])
```

```python
# sobrenome da 4ª pessoa
df['sobrenome'][3]
```

```python
# trazer linha completa
# retorna série
# índice da série retornada -> nome das colunas correspondentes (nome, sobrenome, idade)
# nome da Série (Name) = índice (0)
print(df.iloc[0])
print(type(df.iloc[0]))
```

**## Conclusão: DataFrame é um conjunto de Séries**
 * DataFrame é uma planilha Excel
 * Série é uma coluna do Excel

```python
df.index = [3,2,1,0] #troca de índices
df
```

```python
# [0] não é mais o primeiro elemento
print(df['idade'][0])

# primeiro elemento -> iloc[0]
print(df['idade'].iloc[0])

# loc[0] -> retorna o valor localizado no índice [0] (último elemento)
print(df['idade'].loc[0])
```
